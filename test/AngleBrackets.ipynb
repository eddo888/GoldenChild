```python
import os,sys,re,json,io
from lxml import etree
```

```python
xml='''\
<root>
    <child a="1"/>
    <child>hello</child>
</root>
'''
```

```python
with io.StringIO(xml) as sio:
    tree = etree.parse(sio)
    root = tree.getroot()
    #print(root)

    for child in tree.xpath('/root/child'):
        print(child.tag, child.attrib, child.text)
```

```python
prefix='step'

ns={
    prefix : 'http://www.stibosystems.com/step'
}

text='''\
<STEP-ProductInformation 
    xmlns="http://www.stibosystems.com/step" 
    xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" 
    xsi:schemaLocation="http://www.stibosystems.com/step PIM.xsd" 
    ExportTime="2023-12-20 02:39:48" 
    ExportContext="Context1" 
    ContextID="GL" 
    WorkspaceID="Main" 
    UseContextLocale="false"
>
  <UserTypes>
    <UserType ID="ut1">...</UserType>
  </UserTypes>
</STEP-ProductInformation>
'''
```

```python
xp = '/'.join(['']+list(map(lambda x: f'{prefix}:{x}', [
    'STEP-ProductInformation',
    'UserTypes',
    'UserType'
])))
xp
```

```python
with io.StringIO(text) as sio:
    tree = etree.parse(sio)
    root = tree.getroot()
    print(root.attrib['ExportTime'])

    for ut in tree.xpath(xp, namespaces=ns):
        #print(dir(ut))
        print(ut.nsmap)
        print(ut.tag, ut.attrib, ut.text)   
```
