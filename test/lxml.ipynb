```python
import os, sys, re, json, io
from lxml import etree
```

```python
xml='''\
<root>
    <child a="1"/>
    <child>hello</child>
</root>
'''
```

```python
with io.StringIO(xml) as sio:
    tree = etree.parse(sio)
    root = tree.getroot()
```

```python
for child in tree.xpath('/root/child'):
    #print(dir(ut))
    print(f'{child.tag=}, {child.text=}')
    for a in child.attrib:
        print(f'\t@{a=}, {child.attrib[a]=}')
```

```python
prefix='step'
ns_url='http://www.stibosystems.com/step'

xml=f'''\
<STEP-ProductInformation 
    xmlns="{ns_url}" 
    xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" 
    xsi:schemaLocation="http://www.stibosystems.com/step PIM.xsd" 
    ExportTime="2023-12-20 02:39:48" 
    ContextID="GL" 
    WorkspaceID="Main" 
    UseContextLocale="false"
>
  <UserTypes>
    <UserType ID="ut1">...</UserType>
  </UserTypes>
</STEP-ProductInformation>
'''
```

```python
with io.StringIO(xml) as sio:
    tree = etree.parse(sio)
    root = tree.getroot()
print(root.attrib['ExportTime'])
```

```python
xp = '/'.join(['']+list(map(lambda x: f'{prefix}:{x}', [
    'STEP-ProductInformation',
    'UserTypes',
    'UserType'
])))
xp
```

```python
for child in tree.xpath(xp, namespaces={prefix:ns_url}):
    print(f'{child.nsmap=}')
    print(f'{child.tag=}, {child.text=}')
    for a in child.attrib:
        print(f'\t@ {a} = {child.attrib[a]}')
```
